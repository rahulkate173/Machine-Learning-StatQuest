### Hierarchical Clustering is an unsupervised machine learning algorithm used to group similar data points into clusters. It builds a hierarchy of clusters and is often visualized using a dendrogram.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage,dendrogram
from sklearn.datasets import make_blobs
```

```python
x,_=make_blobs(n_samples=10,centers=3,random_state=0)
```

```python
print(x)
```

```python
z1=linkage(x,method='ward')
z2=linkage(x,method='single')
z3=linkage(x,method='complete')
z4=linkage(x,method='average')
```

```python

fig,ax=plt.subplots(ncols=2,figsize=(10,6))
dendrogram(z1,ax=ax[0])
ax[0].set(xlabel="Sample index",ylabel="Distance",title="Dendrogram (Ward)")
ax[0].legend()
sns.heatmap(z1,ax=ax[1])
```

```python
fig,ax=plt.subplots(ncols=2,figsize=(10,6))
dendrogram(z2,ax=ax[0])
ax[0].set(xlabel="Sample index",ylabel="Distance",title="Dendrogram (Single)")
ax[0].legend()
sns.heatmap(z2,ax=ax[1])
```

```python
fig,ax=plt.subplots(ncols=2,figsize=(10,6))
dendrogram(z3,ax=ax[0])
ax[0].set(xlabel="Sample index",ylabel="Distance",title="Dendrogram (Complete)")
ax[0].legend()
sns.heatmap(z3,ax=ax[1])
```

```python
fig,ax=plt.subplots(ncols=2,figsize=(10,6))
dendrogram(z4,ax=ax[0])
ax[0].set(xlabel="Sample index",ylabel="Distance",title="Dendrogram (average)")
ax[0].legend()
sns.heatmap(z4,ax=ax[1])
```
