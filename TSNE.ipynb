# t-SNE

### Convert the multidimensional data into lower dimensional plot (data) by getting similarity scores and doing hyperparameter tuning

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
digits=load_digits()
x=digits.data
y=digits.target
```

```python
print(x)
```

```python
print(y)
```

```python
print(x.shape)
print(y.shape)
```

```python
tsne=TSNE(n_components=2,perplexity=30,random_state=42)
x_reduced=tsne.fit_transform(x)
```

```python
print(x_reduced)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x_reduced[:,0],x_reduced[:,1],c=y,cmap='tab10',s=15)
plt.title('t-SNE on Digits Dataset')
plt.show()
```
