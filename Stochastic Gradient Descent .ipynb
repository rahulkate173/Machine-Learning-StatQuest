# Stochastic Gradient Descent 

#### For linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# creating data 
np.random.seed(0)
x=np.random.randn(100,1)
true_m,true_c=3,2
y=true_m*x+true_c+np.random.randn(100,1)*0.1 # for some noise 
```

```python
print(x)
print(y)
```

```python
plt.scatter(x,y,label='Data')
plt.plot(x,y,label='line',color='red')
plt.legend()
```

```python
# initilize the random value for slope and intercept
slope=np.random.randn()
intercept=np.random.randn()
```

```python
# setting hyper parameter 
learning_rate=0.001
epochs=100
batch_size=20
```

```python
for epoch in range(epochs):
    index = np.random.permutation(len(x))
    x_shuffled = x[index]
    y_shuffled = y[index]

    for i in range(0, len(x), batch_size):
        x_batch = x_shuffled[i:i+batch_size]
        y_batch = y_shuffled[i:i+batch_size]

        y_pred = slope * x_batch + intercept

        dm = np.mean(2 * x_batch * (y_pred - y_batch))
        dc = np.mean(2 * (y_pred - y_batch))

        slope -= learning_rate * dm
        intercept -= learning_rate * dc

    if epoch % 10 == 0:
        loss = np.mean((slope * x + intercept - y)**2)
        print(f"Epoch {epoch}: Loss = {loss:.4f}")

print(f"\nLearned slope = {slope:.2f}, intercept = {intercept:.2f}")

plt.scatter(x, y, label="Data")
plt.plot(x, slope * x + intercept, color='red', label="Fitted Line")
plt.legend()
plt.title("Linear Regression using Mini-Batch SGD")
plt.show()
```
