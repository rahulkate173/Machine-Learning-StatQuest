# Regression Tree
### Decision tree that are capable for performing regression task 

```python
from sklearn.tree import DecisionTreeRegressor,plot_tree
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
x=iris.data[:,2:]
y=iris.target
reg_tree=DecisionTreeRegressor(max_depth=2) # to avoid overfitting
reg_tree.fit(x,y)
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(ncols=1,nrows=1,figsize=(10,6))
plot_tree(reg_tree,feature_names=iris.feature_names[2:],class_names=iris.target_names,rounded=True,filled=True)
plt.show()
```

```python
# for max depth 3

reg_tree3=DecisionTreeRegressor()
reg_tree3.fit(x,y)

plot_tree(reg_tree3,feature_names=iris.feature_names[2:],class_names=iris.target_names,rounded=True,filled=True)
plt.show()
```

### As we all know that it work to get the minimum least squared residual and set that point as the minimum for the tree
